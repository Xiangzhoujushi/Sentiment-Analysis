```python
import numpy as np
from gensim.models import Word2Vec
from sklearn import linear_model
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import GaussianNB
import gensim
import sys
import os
import pandas as pd
import math
```

```python
sentence_labels = pd.read_csv('sentimentData/labeled_sentences.txt', sep='\t')
sentence_ids = pd.read_csv('sentimentData/datasetSplit.txt', sep=',')
sentence_corpus = pd.read_csv('sentimentData/datasetSentences.txt',sep='\t')
# sentence_corpus
trained_list= sentence_corpus['sentence'].tolist()
train_lists = [x.split() for x in trained_list]
# lists
sentence_df = pd.concat([sentence_ids,sentence_corpus], axis=1).drop(columns = ['sentence_index'])
sentence_df = sentence_df[(sentence_df['splitset_label']==2) | (sentence_df['splitset_label']==1)]
```

```python
# list of sentences
# entence_features = dict()
sentence_list = []
token_list = []
# can split the sentences into single word tokens when we are trying to deal with each word
for index,row in sentence_labels.iterrows():
    li = row['wordstring'].split('|')
    token_list.append(li)
    temp = ' '.join(li)
    sentence_list.append(temp)
sentence_labels['string'] = sentence_list
```

```python
# sentence_labels
token_list= token_list+train_lists
dimension = 30
model = Word2Vec(token_list,size = dimension, window=5,min_count=1)
sentence_labels = sentence_labels[(sentence_labels['splitset_label']==2) | (sentence_labels['splitset_label']==1)]
# print(0)
```

```python
# assign a value for each word
# sentence_ids[sentence_ids['splitset_label']==1]
sentence_features = dict()
maxlength = 56
for index,row in sentence_labels.iterrows():
    matrix = np.zeros(shape = (maxlength,dimension))
    sentence = row['wordstring'].split('|')
    i = 0
    for word in sentence:
        feature= model.wv[word]
        matrix[i] = feature
        i = i+1
    sentence_features[index] = matrix.ravel()
# print(len(sentence_feaures[0])
```

```python
# Compare the learning algorithm

training_set = sentence_features.values()
#separate the trainning and test set
sentence_features
# print(len(sentence_features))
# print(sentence_features.keys())
sentence_labels['features'] = pd.Series(sentence_features)
# sentence_labels.head(10)
sentence_labels["labels"]= [math.floor(x*5)+1 for x in sentence_labels['sentiment']]
test_df = pd.DataFrame(sentence_labels[sentence_labels['splitset_label']==2])
train_df = sentence_labels[sentence_labels['splitset_label']==1]
# sentence_labels
```

```python
#get labels for all data points from strongly negative to strongly positive
#use integers

# 1. Use SGD
# get the trainning data set
clf1 = linear_model.SGDClassifier(loss="hinge",penalty ='l2',alpha = 0.001,tol=0.001)
clf1.fit(train_df["features"].tolist(),train_df["labels"].tolist())
```

```python
# 2. Use SVM
# sentence_labels.head(10)
# train_df["features"].tolist()
clf2 = svm.SVC()
clf2.fit(train_df["features"].tolist(),train_df["labels"].tolist())
```

```python
#3. use neuro-network classifier
clf3 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
clf3.fit(train_df["features"].tolist(),train_df["labels"].tolist())
```

```python
test_df['sgd_result'] = clf1.predict(np.array(test_df['features'].tolist()))
#     temp = ' '.join(row['wordstring'].split('|'))
#     sentence_list.append(temp)
```

```python
test_df['svm_result'] = clf2.predict(np.array(test_df['features'].tolist()))
```

```python
test_df['mlp_result'] = clf3.predict(np.array(test_df['features'].tolist()))
test_df
```

```python
# check the accuracy rate
count1=0
count2=0
count3=0
for index,row in test_df.iterrows():
    if row['sgd_result'] == row['labels']:
        count1+=1
    if row['svm_result'] == row['labels']:
        count2+=1
    if row['mlp_result'] == row['labels']:
        count3+=1
print('The accuracy rate for SGD method',str(float(count1)/len(test_df)))
print('The accuracy rate for SVM method',str(float(count2)/len(test_df)))
print('The accuracy rate for Neuro network method',str(float(count3)/len(test_df)))
```
